# DATA CLEANING FOR REGION & TIMEZONE 

```python
#collecting Region data contain null values
#df2 = Region data contain null values
df1=pd.isnull(df["Region"])
df2=df[df1]
df2
```

```python
df2['Time Zone']
```

```python
#splliting the data using delimeter and extacting(storing) the first string
df2['Time Zone']= df2['Time Zone'].str.split(',').str[0]
df2['Time Zone']
```

```python
#Based on the Time zone replacing all the null values in the Region
df2['Region']=df2["Time Zone"]
df2['Region'] = df2['Region'].replace(['Africa','Asia'],['MENA','APAC'])

df2['Region']
```

```python
# Regions and Timezones are replaced properly

print(df2)
```

```python
df['Time Zone']=df["Region"]
#df['Time Zone']
df['Time Zone'] = df['Time Zone'].replace(['APAC','MENA'],['Asia','Africa'])
df['Time Zone'] 
```

```python
df
```

```python
# df2 values are storing in the main dataframe (df)
df.loc[df2.index, :] = df2[:]
```

```python
df
```

```python

df.isnull().sum()
```

# Data Cleaning for Prefrred language

```python
#using Regex pattern we are storing  the first prefrred language of the person

df['Do you have any preferred language of communication?']=df['Do you have any preferred language of communication?'].replace(to_replace='.*[Ee,]n.*', value= 'English', regex= True)  
df['Do you have any preferred language of communication?']=df['Do you have any preferred language of communication?'].replace(to_replace='.*[Aa,]r.*', value= 'Arabic', regex= True)  
df['Do you have any preferred language of communication?']=df['Do you have any preferred language of communication?'].replace(to_replace='.*[eE,]N.*', value= 'English', regex= True)  
df['Do you have any preferred language of communication?']=df['Do you have any preferred language of communication?'].replace(to_replace='.*[Bb,]a.*', value= 'Bahasa', regex= True)  
df['Do you have any preferred language of communication?']=df['Do you have any preferred language of communication?'].replace(to_replace='.*[Uu,]r.*', value= 'Urdu', regex= True)
df['Do you have any preferred language of communication?']=df['Do you have any preferred language of communication?'].replace(to_replace='.*[Uu,]k.*', value= 'Ukranian', regex= True)  
df['Do you have any preferred language of communication?']=df['Do you have any preferred language of communication?'].replace(to_replace='.*[Tt,]a.*', value= 'Tamil', regex= True)  
df['Do you have any preferred language of communication?']=df['Do you have any preferred language of communication?'].replace(to_replace='.*[Pp,]a.*', value= 'Punjabi', regex= True)  
df['Do you have any preferred language of communication?']=df['Do you have any preferred language of communication?'].replace(to_replace='.*[Hh,]i.*', value= 'Hindi', regex= True)  
```

```python
df
```

# LABEL ENCODING

```python
labelencoder = LabelEncoder()

#df['Do you have any preferred language of communication?'] = labelencoder.fit_transform(df['Do you have any preferred language of communication?'])
df['Country'] = labelencoder.fit_transform(df['Country'])
df['Region'] = labelencoder.fit_transform(df['Region'])
df['Time Zone'] = labelencoder.fit_transform(df['Time Zone'])
```
